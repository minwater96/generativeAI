```python
import tiktoken
enc = tiktoken.encoding_for_model("gpt-4")
```

```python
tokens = enc.encode('자연어 처리가 재밌다')
tokens
```

```python
# 합치기
enc.decode(tokens[:4])
```

```python
tokens = enc.encode('Natural language processing is fun')
tokens
```

```python
enc.decode(tokens[:4])
```
